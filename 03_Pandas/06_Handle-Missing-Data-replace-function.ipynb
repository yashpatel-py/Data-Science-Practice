```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./files/3_weather_data.csv')
df
```

```python
new_df = df.replace(-99999, np.NaN)
new_df
```

```python
new_df = df.replace([-99999, -8888], np.NaN)
new_df
```

```python
new_df = df.replace({
    'temperature': -99999,
    'windspeed': [-99999, -8888],
    'event': '0'
    }, np.NaN)
new_df
```

```python
new_df = df.replace({
    -99999: np.NaN,
    'No Event': 'Sunny'
}, np.NaN)
new_df
```
