### Converting day columns datatype to date datatype

```python
df = pd.read_csv('./files/2_weather_data.csv', parse_dates=['day'])
df.day
```

```python
df.set_index('day', inplace=True)
```

```python
df.head()
```

```python
newdf = df.fillna(0)
newdf
```

```python
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event'
})
new_df
```

```python
new_df = df.fillna(method='ffill')
new_df
```

```python
new_df = df.fillna(method='ffill', axis='columns')
new_df
```

```python
new_df = df.fillna(method='ffill', axis='columns')
new_df
```

```python
new_df = df.fillna(method='bfill', axis='columns')
new_df
```

```python
new_df = df.fillna(method='bfill', limit=1)
new_df
```

```python
new_df = df.interpolate()
new_df
```

## df.interpolate(method='time')

```python
new_df = df.interpolate(method='time')
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')
new_df
```

```python
new_df = df.dropna(thresh=1)
new_df
```
