```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y)
```

format strings

```python
plt.plot(x,y, 'g+')
```

```python
plt.plot(x,y, 'g--')
```

```python
plt.plot(x,y, 'g+-.')
```

```python
plt.plot(x,y, '--r*')
```

```python
plt.plot(x,y, 'rD')
```

keyword arguement

```python
plt.plot(x,y, color='red', marker='D', linestyle="")
```

marker size

```python
plt.plot(x,y, color='red', marker='D', linestyle="", markersize=2)
```

alpha

```python
plt.plot(x,y, color='blue', alpha=0.4)
```
