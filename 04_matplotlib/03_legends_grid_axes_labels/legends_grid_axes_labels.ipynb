```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

setting up the axis label in above chart

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

title in above chart

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

#### adding legend
in order to have legend you need to set label in all the plots as shows in below example


```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")

plt.legend(loc="best", shadow=True)
```

font size

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")

plt.legend(loc="best", shadow=True, fontsize='small')
```
